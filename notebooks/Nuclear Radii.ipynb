```python
plt.hist(np.array(im)[:,:,3].flatten())
plt.yscale('log')
```

```python
axi = plt.imshow((np.array(im)[:,:,3] > 100).astype(int), aspect='auto', cmap=plt.cm.gray, interpolation='nearest')
plt.colorbar()
```

```python
plt.imshow(measure.label((np.array(im)[:,:,3] > 100).astype(int)))
```

```python
regions = measure.regionprops(measure.label((np.array(im)[:,:,3] > 100).astype(int)))
```

```python
areas = [region.area for region in regions]
```

```python
plt.hist(areas, bins=np.linspace(0, 1600, 33))
```

```python
um_per_pix = 0.1333333
areas_um = np.array(areas) * um_per_pix**2
```

```python
plt.hist(areas_um)
```
